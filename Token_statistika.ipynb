```python

import os
import pandas as pd
import numpy as np

folder_path = "/content/drive/MyDrive/DAMSS/Tokenai/Failai"
results = []

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        try:
            df = pd.read_csv(os.path.join(folder_path, filename), sep=';')

            if 'Generated Caption' in df.columns:
                # Skaičiuojame žodžių skaičių
                word_counts = df['Generated Caption'].dropna().apply(lambda x: len(str(x).split()))

                mean_words = np.mean(word_counts)
                var_words = np.var(word_counts)

                results.append({
                    "Failo pavadinimas": filename,
                    "Vidurkis": round(mean_words, 2),
                    "Dispersija": round(var_words, 2)
                })
        except Exception as e:
            print(f"Klaida apdorojant failą {filename}: {e}")

results_df = pd.DataFrame(results)

output_path = os.path.join(folder_path, "token_statistika_news_G.csv")
results_df.to_csv(output_path, sep=';', index=False)

print("Statistika sėkmingai išsaugota faile 'token_statistika_news.csv'")
```
